```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
tau_max = 5.0
tau_r = 1.0
alpha = 1.0
r = np.random.random()
print(r*(np.exp(-alpha*tau_max)-np.exp(-alpha*tau_r)))
tau_2 = (-1/alpha)*np.log(r*(np.exp(-alpha*tau_max)-np.exp(-alpha*tau_r))+np.exp(-alpha*tau_r))
print(tau_2)
```

```python
N = 10000
data = np.zeros(N)
for i in range(N):
    r = np.random.random()
    data[i] = (-1/alpha)*np.log(r*(np.exp(-alpha*tau_max)-np.exp(-alpha*tau_r))+np.exp(-alpha*tau_r))
```

```python
distrib = data
fig, ax = plt.subplots()
ax.hist(distrib, bins=200, range=(0,5), density=False, alpha=0.6)
#ax.hist(values_2, bins=200, range=(0,5), density=True, color="tab:red" ,alpha=0.4)
x = np.linspace(0,5,10000)
ax.plot(x, 1*np.exp(-1*x), color = "tab:orange")
```

```python
N = 10000
data = np.zeros(N)
for i in range(N):
    r = np.random.random()
    data[i] = tau_2 - np.log(1-r)/alpha
```

```python
def exponential_func(x, a, b):
    return a * np.exp(-b * x)
```

```python
data = np.loadtxt("histo.txt")
```
